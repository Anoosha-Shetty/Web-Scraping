```python
current_time = datetime.datetime.now() 
year = str(current_time.year)
month = '%02d' % current_time.month
day = '%02d' % current_time.day
sheetnam = year+month+day

# with pd.ExcelWriter('Jobs_Indeed.xlsx',engine="openpyxl",mode='a') as writer:  
#     Jobs_Indeed.to_excel(writer, sheet_name=sheetnam)
```

```python
pages = [1,2,3,4,5]

title = []
curl = []
sal = []
rat = []
rev = []
comp = []
dat = []
summ = []
loc = []

for pagnum in pages:
    url = 'https://neuvoo.ae/jobs/?k=data&l=united+arab+emirates&date=7d&field=title&company=&source_type=&radius=&from=&test=&iam=&is_category=no&p='+str(pagnum)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    res = soup.find_all(class_ = 'card card__job')
    
    for x in res:
        a1=str(x).find('rel="nofollow"')   #Job Title
        a2=str(x).find('</a>',a1+15)
        fin_a=str(x)[a1+15:a2]
        title.append(fin_a)
        
        b1 = str(x).find('card__job-empname-label"')  # Company Name
        b2 = str(x).find('<',b1+25)
        fin_b = str(x)[b1+25:b2]
        comp.append(fin_b)
        
        c1 = str(x).find('card__job-location"><span>')       # Location
        c2 = str(x).find('</span>',c1+26)
        fin_c = str(x)[c1+26:c2]
        loc.append(fin_c)

        d1 = str(x).find('href="')  # URL for applying for the job
        d2 = str(x).find('"',d1+6)
        fin_d1 = 'https://neuvoo.ae'+str(x)[d1+6:d2]
        fin_d = fin_d1.replace('&amp;','&')
        curl.append(fin_d)

        e1 = str(x).find('class="j-salary">')  # Salary
        if e1 != -1:
            e2 = str(x).find('</div',e1+17)
            fin_e = str(x)[e1+17:e2]
            sal.append(fin_e)
        else:
            sal.append('NA')


    # Summary
        pattern = ['<b>','</b>','</p>']
        i1 = str(x).find('card__job-snippet" dir="auto"')
        i2 = str(x).find('</div>',i1+30)
        fin_i1 = str(x)[i1+30:i2]
        fin_i = re.sub("|".join(pattern),'',fin_i1)
        summ.append(fin_i.strip())
```

```python
data_df = list(zip(comp,title, loc,sal,summ,curl))
Jobs_Neuvoo = pd.DataFrame(data_df, columns = ['Company_Name','Job_Title','Job_Location','Salary',
                                               'Description','URL'])
```

```python
with pd.ExcelWriter('Jobs_Listing.xlsx') as writer:
    Jobs_Indeed.to_excel(writer,sheet_name=sheetnam+'Indeed') 
    Jobs_Neuvoo.to_excel(writer,sheet_name=sheetnam+'Neuvoo') 
```
